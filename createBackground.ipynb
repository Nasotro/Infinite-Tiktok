```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from moviepy.editor import VideoClip
from moviepy.video.io.bindings import mplfig_to_npimage

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Set the limits of the plot
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1, 1)

# Create an array of values for x
x = np.linspace(0, 2*np.pi, 1000)

# Create an empty line object
line, = ax.plot(x, np.sin(x), lw=2)

# Define the animation function
def animate(i):
    line.set_ydata(np.sin(x + i/100))  # update the data
    return line,

# Create the animation
ani = animation.FuncAnimation(fig, animate, frames=100, interval=20)

plt.show()
```

```python
def write_animation(anim, filename, fps=20, duration=10):
    def make_frame(t):
        anim.event_source.stop()
        anim._step()
        return mplfig_to_npimage(fig)

    animation = VideoClip(make_frame, duration=10)
    animation.write_videofile(filename.replace('.mp4','')+'.mp4', fps=20)
```

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

fig = plt.figure()
ax = plt.axes(xlim=(0, 2), ylim=(-2, 2))
line, = ax.plot([], [], lw=2)

def init():
   line.set_data([], [])
   return line,

def animate(i):
   x = np.linspace(0, 2, 1000)
   y = np.sin(2 * np.pi * (x - 0.01 * i))
   line.set_data(x, y)
   return line,

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=200, interval=20, blit=True)
plt.show()

write_animation(anim, 'sine_wave.mp4')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from noise import pnoise2

fig = plt.figure()
im = plt.imshow(np.zeros((100, 100)), cmap='viridis')

def update(frame):
    octaves = 6
    freq = 16.0 * frame / 100.0 if frame > 0 else 0.1
    noise = np.zeros((100, 100))
    for y in range(100):
        for x in range(100):
            noise[x][y] = pnoise2(x / freq, y / freq, octaves=octaves)
    im.set_array(noise)
    return [im]

ani = animation.FuncAnimation(fig, update, frames=100, blit=True)
plt.show()
```

```python
import turtle
import random
from turtle import Screen
screen = Screen() 
screen.setup(width=800,height=600)
#making dots with various different colors on the screen
turtle_array = []
colors = [ "red", "pink", "violet", "yellow",  "light blue", "light green"]
#setting x and y coordinates to the bottom of the screen
y=-200 
x=-200
#using a nested loop
for dot in range (0,10):
    for dot in range (0,10):
        dot = turtle.Turtle()
        dot.penup()
        dot.shape("circle")
        dot.color(random.choice(colors))
        dot.goto(x, y)
        #increasing the y value to shift the dot vertically
        y+=50
        turtle_array.append(dot)
    #increasing the x value to shift the dot horizontally
    #changing the y value back to the bottom of the screen
    y=-200    
    x+=50    
    
    
turtle.done()
```

```python
import turtle
import random
import imageio

# Set up the screen
screen = turtle.Screen()

# Create a turtle object
t = turtle.Turtle()
t.hideturtle()
t.speed(0)

screen.bgcolor("black")

# Function to draw a filled circle
def draw_filled_circle(x, y, radius, color):
    t.hideturtle()
    t.penup()
    t.goto(x, y - radius)
    t.pendown()
    t.color(color)
    t.begin_fill()
    t.circle(radius)
    t.end_fill()

frames = []
canvas = screen.getcanvas()
# Main animation loop
for i in range(200):
    x = random.randint(-300, 300)
    y = random.randint(-300, 300)
    radius = random.randint(10, 100)
    color = (random.random(), random.random(), random.random())
    draw_filled_circle(x, y, radius, color)
    frames.append(imageio.imread(canvas.postscript(colormode='color')))
    imageio.mimsave('animation.gif', frames)

# Close the turtle graphics window
turtle.done()
```
