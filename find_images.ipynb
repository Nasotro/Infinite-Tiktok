```python
import os

def remove_special_chars(text):
    return ''.join(e for e in text.replace(' ', '_') if e.isalnum() and not e == '_')

def dl_images(keywords, path, input_names = False, max_num = 2):
    for i,keyword in enumerate(keywords):
        current_path = os.path.join(path, (remove_special_chars(keyword) if input_names else str(i)))
        google_Crawler = GoogleImageCrawler(storage = {'root_dir': current_path})
        google_Crawler.crawl(keyword = keyword, max_num = max_num, file_idx_offset = 0)

dl_images(['Houdini', 'Eminem'], 'images/tests', max_num=2)
```

```python
import requests
from bs4 import BeautifulSoup
import re

def get_n_image_url(search_term):
    search_term = search_term.replace(' ', '+')
    url = f"https://www.google.com/search?site=&tbm=isch&source=hp&biw=1873&bih=990&q={search_term}"
    headers = {'User-Agent': 'Mozilla/5.0'}

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    a_tags = soup.find_all('a')

    for a_tag in a_tags:
        if a_tag.has_attr('href') and 'imgres' in a_tag['href']:
            return a_tag['href'].split('&amp;imgrefurl=')[1].split('&amp;')[0]

    return None
```
